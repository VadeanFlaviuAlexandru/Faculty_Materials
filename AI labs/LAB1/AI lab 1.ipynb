```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np

#radius of the circle
radius = 1

#side of the equilateral triangle inscribed in the circle
triangle_side = math.sqrt(3) * radius / 2

#number of chords
num_chords = 100

#function to check if a chord is larger than the side of the inscribed equilateral triangle
def chord_larger_than_triangle(chord):
    chord_length = math.sqrt((chord[2] - chord[0])**2 + (chord[3] - chord[1])**2)
    return chord_length > triangle_side

#empty list to store the chords larger than the inscribed equilateral triangle
larger_chords = []

#chords for each case and count the number of larger chords
for i in range(3):
    chords = []
    for j in range(num_chords):
        angle1 = random.uniform(0, 2*math.pi)
        angle2 = random.uniform(0, 2*math.pi)
        x1 = radius * math.cos(angle1)
        y1 = radius * math.sin(angle1)
        x2 = radius * math.cos(angle2)
        y2 = radius * math.sin(angle2)
        chords.append([x1, y1, x2, y2])
    larger_chords_case = [chord for chord in chords if chord_larger_than_triangle(chord)]
    larger_chords += larger_chords_case
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xlim((-1.1*radius, 1.1*radius))
    ax.set_ylim((-1.1*radius, 1.1*radius))
    circle = plt.Circle((0, 0), radius, color='b', fill=False, linewidth=2)
    ax.add_artist(circle)
    for chord in chords:
        linex, liney = [chord[0], chord[2]], [chord[1], chord[3]]
        color = 'r' if chord_larger_than_triangle(chord) else 'k'
        ax.plot(linex, liney, marker='', color=color)
    plt.title('Case {}'.format(i+1))
    plt.show()

#ratio of larger chords to total chords
ratio_larger_chords = len(larger_chords) / (num_chords * 3)

#ratio of larger chords
print('Ratio of chords larger than inscribed equilateral triangle: {:.2f}'.format(ratio_larger_chords))
```
